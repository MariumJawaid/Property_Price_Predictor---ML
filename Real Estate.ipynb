## Creating a Pipeleline

```python
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import StandardScaler
from sklearn.impute import SimpleImputer  #  This was missing

my_pipeline = Pipeline([
    ('imputer', SimpleImputer(strategy="median")),
    ('std_scaler', StandardScaler()),
])
```

```python
housing_num_tr =  my_pipeline.fit_transform(housing)
```

```python
housing_num_tr
```

## Selecting a desired model for Dragon Real Estate

```python
housing_num_tr.shape
```

```python
from sklearn.linear_model import LinearRegression
model = LinearRegression()
model.fit(housing_num_tr , housing_labels)
```

```python
some_data  = housing.iloc[:5]
```

## Evaluating the model

```python
from sklearn.metrics import mean_squared_error
import numpy as np
housing_predictions = model.predict(housing_num_tr)
lin_mse = mean_squared_error(housing_labels ,housing_predictions)
lin_rmse = np.sqrt(lin_mse)
```

```python
lin_mse
```

## Saving Model

```python
import joblib
joblib.dump(model, 'Dragon.pkl')
```

## Testing model on test data

```python
X_test = strat_test_set.drop("MEDV", axis=1)
Y_test = strat_test_set["MEDV"].copy()

# Transform the test features using the pipeline
X_test_prepared = my_pipeline.transform(X_test)

# Make predictions on the prepared test data
final_predictions = model.predict(X_test_prepared)

# Calculate the Mean Squared Error
from sklearn.metrics import mean_squared_error
final_mse = mean_squared_error(Y_test, final_predictions)

# Calculate the Root Mean Squared Error
import numpy as np
final_rmse = np.sqrt(final_mse)

# Display the final RMSE
print("Final RMSE:", final_rmse)
```
